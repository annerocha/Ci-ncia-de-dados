# 3. Modelo Benford:
- Análise Estatística
- Separação Técnica
- Indicador de possível ocorrencia:
 - Descaso
 - Falta de conhecimento
 - Ordem/Concenso Interno
 - Fraude Orçamentária
 - Cenário específico
- Formula do Benford:

[Code Cogs](https://www.codecogs.com/eqnedit.php?latex=\Huge&space;P(digito)&space;=&space;log_{10}(1&space;&plus;&space;\frac{1}{digito}))

$$\Huge P(digito) = log_{10}(1 + \frac{1}{digito})  $$

Probabilidade de aparecer um número cujo primeiro dígito é `1`:

```
import math
```

```
math.log10(1 + (1/1))
```

Probabilidade de aparecer um número cujo primeiro dígito é `2`:

```
d = 3
math.log10(1 + (1/d))
```

## 3.1. Calculando a probabilidade de aparecer de cada primeiro dígito (1 a 9)

```
for i in range(1,10):
  print(math.log10(1 + (1/i)))
```

```
freq_benford = [math.log10(1 + (1/d)) for d in range(1,10)]
freq_benford
```

## 3.2. Benford

### Passo 1: Extrair primeiro dígito

```
def primeiro_digito(n):
  return str(n)[0]
```

```
primeiro_digito(563)
```

Extraindo  o primeiro dígito de cada número da coluna

```
digs = df_yt['views'].apply(primeiro_digito)
digs
```

### Passo 2: calculando a quantidade dos primeiros dígitos 

Teste de mesa:

```
digs = df_yt['views'].apply(primeiro_digito)
digs.value_counts(normalize=True)
```

Criando a função:

```
def freq_real(coluna):
  digitos = coluna.apply(primeiro_digito)
  digitos = digitos[digitos != '0']
  freq = digitos.value_counts(normalize=True)
  return freq
```

```
freq_real(df_yt['views'])
```

## Passo 3: criando um dataframe com colunas real e esperada

Teste de mesa

```

coluna = df_yt['views']
df_benford = pd.DataFrame()
df_benford['Real'] = freq_real(coluna)
df_benford['Benford'] = [math.log10(1 + (1/d)) for d in range(1, 10)]
df_benford.plot.bar(figsize=(16,6));
```

Definindo a função

```
def benford(coluna: pd.Series):
  df_benford = pd.DataFrame()
  df_benford['Real'] = freq_real(coluna)  
  df_benford['Benford'] = [math.log10(1 + (1/d)) for d in range(1, 10)]
  df_benford.plot.bar(figsize=(16,6));
  return df_benford
```

Testando a função

```
benford(df_aux_emerg['Qtd'])
```

```
benford(df_yt['views']);
```
